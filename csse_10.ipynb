```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import math

# Set random seed for reproducibility
np.random.seed(1000)

# Define the function to learn
function_to_learn = lambda x: np.cos(x) + 0.1 * np.random.randn(*x.shape)

# Define hyperparameters
layer_1_neurons = 10
NUM_points = 1000
batch_size = 100
NUM_EPOCHS = 1500

# Generate training and validation data
all_x = np.float32(np.random.uniform(-2 * math.pi, 2 * math.pi, (1, NUM_points))).T
np.random.shuffle(all_x)
train_size = 900

x_training = all_x[:train_size]
y_training = function_to_learn(x_training)

x_validation = all_x[train_size:]
y_validation = function_to_learn(x_validation)

# Visualize the training and validation data
plt.figure()
plt.scatter(x_training, y_training, c='blue', label='Train')
plt.scatter(x_validation, y_validation, c='pink', label='Validation')
plt.legend()
plt.show()

# Build the model using the Keras API
model = tf.keras.Sequential([
    tf.keras.layers.Dense(layer_1_neurons, activation='sigmoid', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
history = model.fit(
    x_training, y_training,
    validation_data=(x_validation, y_validation),
    batch_size=batch_size,
    epochs=NUM_EPOCHS,
    verbose=1
)

# Plot the training and validation loss
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

```
